```python
plt.figure(figsize=(10, 6))
sns.histplot(train_data['SalePrice'], kde=True)
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
skew = train_data['SalePrice'].skew()
kurt = train_data['SalePrice'].kurt()
print(f'Skewness of SalePrice: {skew}')
print(f'Kurtosis of SalePrice: {kurt}')
```

Fromt the information we have got so far  
**our target data is positive skewed with high kurtosis(long tailed)**  
**The high skew and kurt shows the presense of outliers in ourt data**

```python
features_to_visualize = ['LotArea', 'YearBuilt', 'TotalBsmtSF', 'GrLivArea']
train_data[features_to_visualize].hist(bins=15, figsize=(15, 6), layout=(2, 2))
```

```python
correlation_matrix = train_data.corr()
saleprice_correlation = correlation_matrix['SalePrice'].sort_values(ascending=False)
# Convert the correlation series to a DataFrame
saleprice_correlation_df = saleprice_correlation.to_frame()

# Visualizing the correlation with SalePrice
plt.figure(figsize=(10, 12))
heatmap = sns.heatmap(saleprice_correlation_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
heatmap.set_title('Features Correlating with SalePrice', fontdict={'fontsize':18}, pad=16)
plt.show()
```
