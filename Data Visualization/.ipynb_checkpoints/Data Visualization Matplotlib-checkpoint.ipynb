```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
spotify_data = pd.read_csv('./spotify.csv')
spotify_data.head()
```

```python
spotify_data.isnull().sum()
```

```python
# Since here nan represent the 0 views at that time
spotify_data.fillna(0, inplace=True)
```

```python
spotify_data.info()
```

From the data information we can see that our date object is in object format for us to use it as date let's change it to date time

```python
spotify_data['Date'] = pd.to_datetime(spotify_data['Date'])
```

```python
plt.figure(figsize=(10, 6))

for column in spotify_data.columns[1:]:
    plt.plot(spotify_data['Date'], spotify_data[column], label=column)
    
# Adding Labels and Title
plt.xlabel('Date')
plt.ylabel('Streams')
plt.title('Song Streams Over Time')
plt.legend()

plt.show
```

```python
plt.plot(spotify_data['Date'], spotify_data['Shape of You'], label = spotify_data['Shape of You'])
plt.show()
```

```python
plt.plot(spotify_data['Date'], spotify_data['Despacito'])
plt.show()
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool
```

```python
output_notebook()

p = figure(title="Shape of You - Stream Over Time", 
           x_axis_label = 'Date', 
           y_axis_label='Streams',
           x_axis_type='datetime',
           width=800,
           height=400
          )

p.line(spotify_data['Date'], spotify_data['Shape of You'],
       legend_label="Shape of You",
       line_width=2
      )

hover = HoverTool()
hover.tooltips=[
    ('Date', '@x{%F}'),
    ('Streams', '@y'),
]
hover.formatters={'@x': 'datetime'}
p.add_tools(hover)

# Show the results
show(p)
```

```python
total_views_to_date = np.zeros(len(spotify_data['Shape of You']))
```

```python
for i in range(len(spotify_data['Shape of You'])):
    total_views_to_date[i] = np.sum(spotify_data['Shape of You'][:i+1])
```

```python
spotify_data['Shape of You'][:1]
```

```python
spotify_data['Total Views Shape of You'] = spotify_data['Shape of You'].cumsum()
```

```python
output_notebook()

p = figure(title="Shape of You - Stream Over Time", 
           x_axis_label = 'Date', 
           y_axis_label='Streams',
           x_axis_type='datetime',
           width=900,
           height=600
          )

p.line(spotify_data['Date'], spotify_data['Total Views Shape of You'],
#        legend_label="Shape of You",
       line_width=2
      )

hover = HoverTool()
hover.tooltips=[
    ('Date', '@x{%F}'),
    ('Streams', '@y'),
]
hover.formatters={'@x': 'datetime'}
p.add_tools(hover)

# Show the results
show(p)
```
