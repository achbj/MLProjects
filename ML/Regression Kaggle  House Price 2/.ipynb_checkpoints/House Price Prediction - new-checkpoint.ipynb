```python
X = train.drop(columns=['SalePrice'], axis=1)
y = np.log1p(train['SalePrice'])
```

## Pipeline for data imputation and onehotencoding and normalization

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
def get_preprocessor(df):
    # pipeline for numeric column 
    numeric_transformer = make_pipeline(
        SimpleImputer(strategy='median'),
        StandardScaler()
    )
    
    # pipeline for categorical column
    categorical_transformer = make_pipeline(
        SimpleImputer(strategy='most_frequent'),
        OneHotEncoder(handle_unknown='ignore')
    )
    
    # Separate numerical and categorical column
    numeric_columns = df.select_dtypes(exclude='object').columns
    categorical_columns = df.select_dtypes(include='object').columns
    
    # combine pipeline using columntransformer
    preprocessor = ColumnTransformer(
        transformers= [
            ('num', numeric_transformer, numeric_columns),
            ('cat', categorical_transformer, categorical_columns)
        ]
    )
    
    return preprocessor
```

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

# Existing preprocessor
preprocessor = get_preprocessor(X)

# set up the feature selector
selector = RFECV(estimator=LinearRegression(), step=1, cv=5)

# Create a pipeline that first applies your preprocessor, and then the selector
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('selector', selector),
    ('model', LinearRegression())
])

# Fit the pipeline
pipeline.fit(X, y)

selected_features = np.where(pipeline.named_steps['selector'].support_)[0]
```

```python
# Get the transformed column names after preprocessing
numeric_columns = X.select_dtypes(exclude='object').columns
categorical_columns = preprocessor.named_transformers_['cat']['onehotencoder'].get_feature_names_out(X.select_dtypes(include='object').columns)

transformed_columns = np.concatenate([numeric_columns, categorical_columns])

# Get the mask of selected features
selected_mask = pipeline.named_steps['selector'].support_

# Get the names of the selected features
selected_columns = transformed_columns[selected_mask]
```
