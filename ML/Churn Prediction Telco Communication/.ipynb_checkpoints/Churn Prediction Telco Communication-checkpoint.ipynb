```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

# Data Preprocessing

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'].astype
```

The column `TotalCharges` is in object Type so we need to change it to float to play with this column


```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.columns
```

```python
from sklearn.preprocessing import OneHotEncoder

# list of the columns to encode
columns_to_encode = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'Churn']

# Extract the columns to encode
data_to_encode = df[columns_to_encode]

# Initialize the OneHotEncoder
encoder = OneHotEncoder()

#Fit and transform the data
encoded_data = encoder.fit_transform(data_to_encode)

# Convert the encoded data to DataFrame
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(columns_to_encode))

# Concat the encoded DataFrame with the original dataframe
df_en = pd.concat([df, encoded_df], axis=1)
df_en.drop(columns=columns_to_encode, inplace=True)
```

# Data Visualization

Let's visualize the ratio of the churned and not churned users ratio first and their ratio to other features

```python
# Count the number of churned and non-churned customers
churn_counts = df_en[['Churn_Yes', 'Churn_No']].sum()

# Labels for the pie chart
labels = ['Churned', 'Not Churned']

# Colors for the piechart
colors = ['red', 'blue']

# Explode a slice
explode = (0.1, 0)

# Create a piechart
plt.figure(figsize=(6,6))
plt.pie(churn_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)

# Add a legend in the right top corner
plt.legend(labels, loc="upper right")

plt.title('Churn Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you already have your DataFrame df_en with 'Churn_Yes', 'Churn_No', 'gender_Female', and 'gender_Male' columns

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Labels and colors for the pie charts
labels = ['Female', 'Male']
colors = ['lightcoral', 'lightskyblue']

# Explode sections for emphasis
explode = (0.1, 0)

# Calculate the gender distribution for Churned customers
churned_gender_counts = df_en[df_en['Churn_Yes'] == 1][['gender_Female', 'gender_Male']].sum()

# Calculate the gender distribution for Not Churned customers
not_churned_gender_counts = df_en[df_en['Churn_No'] == 1][['gender_Female', 'gender_Male']].sum()

# Pie chart for Churned
axes[0].pie(churned_gender_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
axes[0].set_title('Churned Customers')

# Pie chart for Not Churned
axes[1].pie(not_churned_gender_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
axes[1].set_title('Not Churned Customers')

# Add a legend
axes[1].legend(labels)

# Set a common title for the subplots
fig.suptitle('Gender Distribution in Churned and Not Churned Customers')

# Display the subplots
plt.show()
```
