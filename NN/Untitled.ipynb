```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("Deci/DeciLM-7B")
model = AutoModelForCausalLM.from_pretrained("Deci/DeciLM-7B")
```

```python
def GetResponse(prompt):
    inputs = tokenizer.encode(prompt, return_tensors='pt')
    outputs = model.generate(inputs, max_new_tokens=110, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id, do_sample=True, top_p=0.95)
    res = tokenizer.decode(outputs[0])
    
    return res
```

```python
prompt="What is the capital of Nepal?"
res = GetResponse(prompt)
```
